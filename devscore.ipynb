```python
import requests
import json
import pandas as pd
import time


URL = "https://api.coingecko.com/api/v3/"
payload = {"vs_currency":"usd"}


r = requests.get(URL+"coins/markets", params=payload)
d = r.json()
df_top100 = pd.DataFrame(d)
df_top100.head()

#Going to fetch the top 100 cryptos
#To be kind to server, let's sleep 0.5s after each call :)
payload = {"localization":"false"}
market_data = []
developer_data = []
for i in df_top100['id']:
    try:
        r = requests.get(URL+"coins/"+i, params=payload)
        data = r.json()
        print(f"#{data['market_cap_rank']}: <Fetched> {data['name']}")
        market_data.append(data)
        developer_data.append(data['developer_data'])
        time.sleep(0.5)
    except Exception as e:
        print(i)
        print(e)
    
df_top100_mdata = pd.DataFrame(market_data)
df_top100_ddata = pd.DataFrame(developer_data)

df_top100_mdata.to_csv("top100marketdata.csv")
df_top100_ddata.to_csv("top100devdata.csv")
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd


df_m = pd.read_csv("top100marketdata.csv")
df_d = pd.read_csv("top100devdata.csv")
df_d.head()
df_m.head()

df = pd.merge(df_m, df_d)

df_sorted = df.sort_values(by=['commit_count_4_weeks'], ascending=False)
df_top = df_sorted.head(20)
df_top.head(20)
df_top['name'] = np.where(df_top['name'] =="Synthetix Network Token", "Synthetix\nNetwork Token", df_top['name'])
plt.figure(dpi=600,figsize=(9,12))
sns.despine()
sns.set_context('paper')
# sns.set_style("white")
pal = sns.light_palette("Blue", len(df_top), reverse=True)
# plt.tight_layout()
# plt.subplots_adjust(top=.95)
ax = sns.barplot(x='name', y='commit_count_4_weeks', data = df_top, palette=pal, dodge=False)

right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
ax.set_ylabel("4 weeks github commit count")
# labels = [item.get_text() for item in ax.get_yticklabels()]
# ax.set_yticklabels(labels, rotation='horizontal', fontsize=6)


labels = [item.get_text() for item in ax.get_xticklabels()]
for i in range(len(labels)):
    labels[i] = f"{df_top.iloc[i]['name']} [#{df_top.iloc[i]['market_cap_rank']}]"

ax.set_xticklabels(labels)
ax.set_xticklabels(labels, rotation='vertical', fontsize=9)

i = 0
for p,val in zip(ax.patches, df_top['commit_count_4_weeks']):
    ax.annotate("  "+str(val), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'bottom', 
                   xytext = (0, 9), 
                   textcoords = 'offset points',
                rotation = 'vertical',
                size = 7
               )
    i+=1



plt.savefig('filename2.png')
```
